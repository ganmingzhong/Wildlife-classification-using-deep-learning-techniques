```python
#pillow 
def hsv_value(img,factor):
    hsv_image = img.convert('HSV')
    hsv_array = np.array(hsv_image)
    
    
    hsv_array[:, :, 2] = np.clip(hsv_array[:, :, 2] * factor, 0, 255)
    
    #convert back to rgb
    hsv_image = Image.fromarray(hsv_array, 'HSV')
    adjusted_image = hsv_image.convert('RGB')
    #adjusted_image.save('adjusted_image.jpg')
    #adjusted_image.show()

    return adjusted_image

def hsv_hue(img,factor):
    hsv_image = img.convert('HSV')
    hsv_array = np.array(hsv_image)
    
    
    hsv_array[:, :, 0] = np.clip(hsv_array[:, :, 0] * factor, 0, 255)
    
    #convert back to rgb
    hsv_image = Image.fromarray(hsv_array, 'HSV')
    adjusted_image = hsv_image.convert('RGB')
    #adjusted_image.save('adjusted_image.jpg')
    #adjusted_image.show()

    return adjusted_image

def hsv_saturation(img,factor):
    hsv_image = img.convert('HSV')
    hsv_array = np.array(hsv_image)
    
    
    hsv_array[:, :, 1] = np.clip(hsv_array[:, :, 1] * factor, 0, 255)
    
    #convert back to rgb
    hsv_image = Image.fromarray(hsv_array, 'HSV')
    adjusted_image = hsv_image.convert('RGB')
    #adjusted_image.save('adjusted_image.jpg')
    #adjusted_image.show()

    return adjusted_image
```

```python
def rotate(img):
    degree = random.randint(30, 330)
    img_rotate = img.rotate(degree)
    return img_rotate

def translation(image):
    height, width = image.shape[:2] 
  
    fraction_height, fraction_width = height / 6, width / 6
      
    T = np.float32([[1, 0, fraction_width], [0, 1, fraction_height]]) 
    print(T)  
    # We use warpAffine to transform 
    # the image using the matrix, T 
    img_translation = cv2.warpAffine(image, T, (width, height)) 
    return img_translation

def scale(img):
    size = random.randint(100, 500)
    img_scale = img.resize((size, size))
    return img_scale
```

```python
def flip_updown(img):
    updown_img=img.transpose(Image.FLIP_TOP_BOTTOM)
    return updown_img

def flip_leftright(img):
    leftright_img=img.transpose(Image.FLIP_LEFT_RIGHT)
    return leftright_img

def perspective(img):
    height, width,c = img.shape
    a= width/3
    b= width - width/3
    pts1 = np.float32([[0,0],[width,0],[0,height],[width,height]])
    pts2 = np.float32([[a,0],[width,0],[0,height],[b,height]])
    
    matrix = cv2.getPerspectiveTransform(pts1,pts2)
    result = cv2.warpPerspective(img,matrix,(width,height))
    return result
```

# Shear Mosaic Blur


```python
def shear_image(image, shear_factor):
    shear_matrix = [1, shear_factor, 0, 0, 1, 0]
    sheared_image = image.transform(image.size, Image.AFFINE, shear_matrix)
    return sheared_image
```

# Gaussian Bright Contrast

```python
def add_gaussian_noise(image, mean=0, stddev=0):
    # Convert image to numpy array
    img_array = np.array(image)

    # Generate random Gaussian noise with the same shape as the image
    noise = np.random.normal(mean, stddev, img_array.shape)

    # Add noise to the image
    noisy_image = np.clip(img_array + noise, 0, 255).astype(np.uint8)

    # Convert noisy image back to PIL Image
    noisy_image = Image.fromarray(noisy_image)

    return noisy_image
```
